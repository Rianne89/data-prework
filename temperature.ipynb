As changing the cooling system is expensive and you are an excellent data analyst, you can't make a decision without basis.

## Tasks
The temperatures measured throughout the 24 hours of a day are:

```python
temperatures_C = [33, 66, 65, 0, 59, 60, 62, 64, 70, 76, 80, 81, 80, 83, 90, 79, 61, 53, 50, 49, 53, 48, 45, 39]
```

The first element of the list is the temperature at 12am, the second element is the temperature at 1am, and so on. 

The company has decided that if one of the following events occurs, then the cooling system needs to be replaced for a new one to avoid damaging the processor.
* More than 4 temperatures are greater than or equal to 70¬∫C.
* Any temperature is above 80¬∫C.
* The average temperature exceeds 65¬∫C.

Follow the steps so that you can make the decision.

#### 1. Find the minimum temperature of the day and store it in a variable.

```python
Min_temp = min(temperatures_C)
print(Min_temp)
```

#### 2. Find the maximum temperature of the day and store it in a variable.

```python
Max_temp = max(temperatures_C)
print(Max_temp)
```

#### 3. Create a list with the temperatures that are greater than or equal to 70¬∫C. Store it in a variable.

```python
Temp_70 = []

for i in temperatures_C:
    if i >= 70:
        Temp_70.append(i)
    
print(Temp_70)
```

#### 4. Find the average temperature of the day and store it in a variable.

```python
Ave_Temp = (sum(temperatures_C)/len(temperatures_C))
print(Ave_Temp)
```

#### 5. Imagine that there was a sensor failure at 3am and the data for that specific hour was not recorded. How would you estimate the missing value? Replace the current value of the list at 3am for an estimation. 

```python
temperatures_C.insert(3,60)
temperatures_C.remove(0)
print(temperatures_C)
```

#### 6. Bonus: the maintenance staff is from the United States and does not understand the international metric system. Help them by converting the temperatures from Celsius to Fahrenheit.
To know more about temperature conversion check this [link](https://en.wikipedia.org/wiki/Conversion_of_units_of_temperature).

**Formula**: 

$F = 1.8 * C + 32$

```python
temperatures_F = []

for i in temperatures_C:
    fahrenheit = ((1.8*i)+32)
    temperatures_F.append(fahrenheit)

print(temperatures_F)
```

#### 7. Make a decision!
Now it's time to make a decision taking into account what you have seen until now. 

Remember that if one of the following events occurs, then the cooling system needs to be replaced for a new one to avoid damaging the processor.
* More than 4 temperatures are greater than or equal to 70¬∫C.
* Any temperature is above 80¬∫C.
* The average temperature exceeds 65¬∫C.

#### To make your decision, check if any of the three conditions above is met. You might need to use some of the variables you created in steps 1 to 6. Print a message to show if the cooling system needs to be changed or not.

```python
print("The cooling system needs to be replaced for a new one, because", (len(Temp_70)), "temperatures are greater than or equal to 70¬∫C")
```

```python
Ave_temp_C = sum(temperatures_C)/len(temperatures_C)
Ave_temp_F = sum(temperatures_F)/len(temperatures_F)

print(Ave_temp_C)
print(Ave_temp_F)

print("The relation between both average values is the formula: ùêπ=1.8‚àóùê∂+32")
```

#### 5. Find the standard deviation of the temperature lists (¬∫C and ¬∫F). What is the relation between both standard deviations?

```python
Deviation_from_mean_C = []

for i in temperatures_C:
    deviation = (i - Ave_temp_C)
    Deviation_from_mean_C.append(deviation)

square_deviation_C = []

for i in Deviation_from_mean_C:
    deviation2 = (i * i)
    square_deviation_C.append(deviation2)

Variance_C = sum(square_deviation_C)/len(square_deviation_C)

print("Standard deviation Celsius:",(Variance_C**0.5))

Deviation_from_mean_F = []

for i in temperatures_F:
    deviation_F = (i - (Ave_temp_F))
    Deviation_from_mean_F.append(deviation_F)

square_deviation_F = []

for i in Deviation_from_mean_F:
    deviation_s2 = (i * i)
    square_deviation_F.append(deviation_s2)

Variance_F = (sum(square_deviation_F)/len(square_deviation_F))

print("Standard deviation Fahrentheit:",(Variance_F**0.5))

print("The relation between both average values is the formula: ùêπ=1.8‚àóùê∂+32")
```
